```python
import requests                                           
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.bbc.co.uk/newsround/59402668'
comment =[]
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
# Audience Review
audience_reviews=soup.findAll("div", {"class": "comment__text"})
for review in audience_reviews:
  #print(review.text)
  comment.append(review.text)
dct={'comment':comment}
df= pd.DataFrame.from_dict(dct)
df.head(10)
```
