```python
import requests                                           
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Blog sites
url='https://www.theguardian.com/film/2021/dec/18/spider-man-no-way-home-review-more-is-more'
comment =[]
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
# Audience Review
audience_reviews=soup.findAll("div", {"class": "dcr-2rh10c"})
for review in audience_reviews:
  print(review.text)
  comment.append(review.text)
dct={'comment':comment}
df= pd.DataFrame.from_dict(dct)
df.head()
```
