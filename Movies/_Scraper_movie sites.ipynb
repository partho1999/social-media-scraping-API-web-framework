```python
import requests                                           
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#for IMDb 
url = 'https://www.imdb.com/title/tt10872600/reviews?ref_=tt_urv'

r = requests.get(url)
#print(r.text)

soup = BeautifulSoup(r.text, 'html.parser')

#print(soup.title.text)

reviews = soup.find_all("div", {"class": "lister-list"})
#print(reviews)
title = []
rate = []
content = []

for item in reviews:

  tiltes = (item.find_all('a' , {"class":"title"}))
  rating = (item.find_all('div' , {"class":"ipl-ratings-bar"}))
  contents=(item.find_all('div' , {"class":"text show-more__control"}))
  #print(contents)

  for t in tiltes:
    title.append(t.text.strip())
  for r in rating:
    rate.append(r.text.replace('/10','').strip())
  for line in contents:
    content.append(line.text.strip()) 

dct={'title':title,'rate':rate,'content':content}

df= pd.DataFrame.from_dict(dct,orient='index')
df = df.transpose()
df.head(26)
df.to_csv('IMDB.csv')
```

```python
#rotten tomatoes
url='https://www.rottentomatoes.com/m/spider_man_no_way_home'
comment =[]
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
# Audience Review
audience_reviews=soup.findAll("ul", {"class": "mop-audience-reviews__reviews-wrap clearfix"})
for review in audience_reviews:
  #print(review.text)
  comment.append(review.text)
dct={'comment':comment}
df= pd.DataFrame.from_dict(dct)
df.head()
```
